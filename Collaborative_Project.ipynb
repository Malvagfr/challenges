## Data exploration, extration and wrangling

```python
#library importing
import pandas as pd
import requests
import bs4
import re
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Nobel_laureates'
```

```python
response = requests.get(url)
```

```python
html=response.content
```

```python
parsed_html=bs4.BeautifulSoup(html,"html.parser")
```

```python
table = parsed_html.find_all("table",{'class':'wikitable'})
```

```python
first_row=table.find_all("th")
```

```python
titles = [first_row[title].text.replace('\n','') for title in range(int(len(first_row)/2))]
titles
```

```python
rows=table.find_all("tr")
```

```python
[i.text for i in row][1::2]
```

```python
body=[[i.text for i in row][1::2] for row in rows]
body
```

```python
df = pd.DataFrame(body)
df
```

```python
df=df.iloc[1:-1]
df
```

```python
df.columns=titles
df
```
