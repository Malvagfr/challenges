# Challenge 1

Warm up
We want to create bins of movies according to the number of votes they've received. For that matter, we will create a new column named 'bin' which will tag every movie as follow:

- From 0 to 999 ==> 1
- From 1000 to 9999 ==> 2
- From 10000 to 99999 ==> 3
- From 100000 to 999999 ==> 4
- More than 1000000 ==> 5

```python
import pandas as pd
```

```python
data = pd.read_csv('./data/input/IMDB-Movie-Data.csv')
```

```python
data.head()
```

```python
def bin_calc(x):
    if x>=0 and x<=999:
        y=1
    elif x>=1000 and x<=9999:
        y=2
    elif x>=10000 and x<=99999:
        y=3   
    elif x>=100000 and x<=999999:
        y=4
    elif x>=1000000:
        y=4
    return y
```

```python
data['bin']=data['Votes'].apply(bin_calc)
```

# Challenge 3
Using the lambda

We want to create a new ranking where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.

```python
def func_bin_2(x,y):
    if 'Thriller' in x:
        y+=1
    elif 'Comedy' in x:
        y-=1
    return y
```

```python
data['bin_2'] = data.apply(lambda x: func_bin_2(x['Genre'],x['bin']), axis=1)
```

# Challenge 4. 
    
Now the real stuff
    
We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers

```python
#x is Actors and y is Votes
def prime_number(x,y):
    #calculating ASCII
    num=0
    for i in range(len(x)):
        num += ord(x[i])
    #calculating number
    n=num/y
    
    
    #calculating if number is prime

    # 0, 1, even numbers greater than 2 are NOT PRIME
    if (n>=0 and n<=1) or (n % 2 == 0 and n > 2):
        return False
    else:
        # Not prime if divisable by another number less
        # or equal to the square root of itself.
        # n**(1/2) returns square root of n
        for i in range(3, int(n**(1/2))+1, 2):
            if n%i == 0:
                return False
        return True
```

```python
data['Prime_div'] = data.apply(lambda x: prime_number(x['Actors'],x['Votes']), axis=1)
```

# Challenge 6. Freaky bonus

We want to know which movies might have hidden paterns in their description.
A way to know that is finding those movies which the sum of all numeric values of the string 
description hash (SHA256) are between their revenue and their number of votes.

```python
def hidden_paterns(row):
    #taking numeric values
    numbers = []
    for word in row['Description'].split():
        if word.isdigit():
            numbers.append(int(word))
    #calculating the sum        
    numbers_sum=0
    for number in numbers:
        numbers_sum+=number
    #checking whether the sum is between votes and revenue
    if (row['Votes'] <= numbers_sum and numbers_sum <= row['Revenue (Millions)']) or (row['Revenue (Millions)'] <= numbers_sum and numbers_sum <= row['Votes']):
        return True
    else:
        return False
```

```python
data['hidden_paterns']=data.apply(hidden_paterns,axis=1)
```
