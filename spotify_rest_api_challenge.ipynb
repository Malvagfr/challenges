## Get artists top tracks

```python
def get_top_tracks(artist_id):
    base_url = 'https://api.spotify.com/v1'
    resource = '/artists/'
    headers = {'Authorization': 'Bearer {token}'.format(token=access_token)}
    market = '?market=ES'
    get_type='/top-tracks'
    pulls_response_top_tracks = requests.get(base_url + resource + artist_id + get_type + market,headers=headers)
    pulls_json_top_tracks = pulls_response_top_tracks.json()
    pulls=pulls_json_top_tracks['tracks']
    df_pulls_top_tracks = pd.DataFrame(pulls_json_top_tracks['tracks'])[['href','id','name']]
    df_pulls_top_tracks['id artist']=artist_id
    return df_pulls_top_tracks
```
